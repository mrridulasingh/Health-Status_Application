```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("breast_cancer_data.csv")
df
```

```python
df.info()
```

```python
df.diagnosis.value_counts()
```

```python
df=pd.get_dummies(data=df, columns=['diagnosis'], drop_first=True)
df
```

```python
df=df.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(25,15))
sns.heatmap(df.corr(), annot=True, linewidths=0.5)
```

```python
df.columns
```

```python
df=df.drop(['radius_mean', 'perimeter_mean', 'area_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'smoothness_se',
        'symmetry_se','area_worst', 'fractal_dimension_se','perimeter_se', 'area_se', 'radius_worst','perimeter_worst'], axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('diagnosis_M', axis=1)
y=df['diagnosis_M']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=45)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
pred=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
score=accuracy_score(pred, y_test)
score.mean()
```

```python
confusion_matrix=confusion_matrix(pred, y_test)
confusion_matrix
```

```python
report=classification_report(pred, y_test)
print(classification_report)
```
